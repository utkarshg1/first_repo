# Show the total bill by day

```python
a = (
    df.groupby(by = "day")
    .agg({"total_bill": "sum"})
    .sort_values(by = "total_bill", ascending=False)
)
a
```

```python
a.plot(kind="bar")
```

```python
a.plot(kind="barh")
```

# Show sum of total bill and sum of tip by day

```python
b = (
    df.groupby(by = "day")
    .agg({"total_bill": "sum", "tip": "sum"})
    .sort_values(by = "total_bill", ascending=False)
)
```

```python
b
```

```python
b.plot(kind="bar")
```

# Show average bill by day and time

```python
c
```

```python
c.plot(kind="barh")
```

# Create a pivot table , show total bill by day and time

```python
d = pd.pivot_table(
    data=df,
    index="day",
    columns="time",
    values = "total_bill",
    aggfunc = "sum"
)
```

```python
d
```

```python
d.plot(kind="bar")
```

# Show the pivot table by day and size show mean of tip

```python
e
```

# Show total profit by Sub-Category

# Show top 3 subcategory by total profit from East region
