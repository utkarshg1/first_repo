### Assignment - For given loan amount of 12 lakhs for 8 years at 8.5%p.a. calculate the EMI for this loan amount

```python
# Take P, N, R as input from use
P = 12_00_000
N = 8
R = 8.5

# Convert number of years to month and rate of intrest to %per month
n = N*12 
r = R/1200

# Calculating emi
x = (1 + r)**n
emi = P*r*x/(x-1)

# Total amount to be paid
amt = emi * n

# Intrest
I = amt - P

# Printing all above results
print(f"Prinicipal = {P} INR, Years = {N} , Rate of Intrest = {R} %p.a.")
print(f"EMI = {emi:,.0f} INR")
print(f"Total Amount = {amt:,.0f} INR")
print(f"Intrest = {I:,.0f} INR")
```

### Check wheter user is eligible to vote

```python
# Take age as input from user
age = int(input("Please enter age in years :"))

# Check whether age >= 18
if age >= 18:
    print(f"At age {age} years , YES You are eligible to vote")
    print("====================")
else:
    print(f"At age {age} years, You are NOT eligible to vote")
    print("*********************")
```

### Check wheter a given number is Divisible by BOTH 7 AND 3

```python
num = int(input("Please enter a positive number :"))

if num%7 == 0 and num%3 == 0:
    print("The given number is DIVISBILE by BOTH 7 AND 3")
else:
    print("The given number is NOT Divisible by Either 3 or by 7 or both")
```
